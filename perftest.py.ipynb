```julia
import pandas
import numpy
import matplotlib.pyplot as plt
```

```julia
x_min = -2
x_max = 5
count = 100_000_000

low = 0.2
high = 2.0
size = count

num_points_to_plot = 1_000
```

```julia
x = numpy.linspace(x_min, x_max, count)
random_noise = numpy.random.uniform(low, high, size)
y = x**2 - 2*x + random_noise

df = pandas.DataFrame({'x':x, 'y':y})
```

```julia
# Sample and sort
df_sampled = df.sample(n=num_points_to_plot, random_state=42)
df_sampled_sorted = df_sampled.sort_values(by='x').reset_index(drop=True)
```

```julia
plt.figure( figsize=(10,6) )
plt.plot( df_sampled_sorted.x, df_sampled_sorted.y,
    label=r'$y(x) = x^2 - 2x + \text{random}$', )
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.title(r'Plot of $y(x) = x^2 -2x + \text{random}$', fontsize=18)
plt.grid(True)
plt.legend()

x_curve = numpy.linspace(x_min, x_max, 500)
y_curve = x_curve**2 - 2*x_curve

plt.plot(x_curve, y_curve, color='red', linestyle='--', label=r'Original Function ($y=x^2-2x$)')


# --- Labels and Title with LaTeX ---
plt.xlabel(r'$x$') # Using raw string and math mode for 'x'
plt.ylabel(r'$y$') # Using raw string and math mode for 'y'
plt.title(r'Plot of Sampled Data from $y(x) = x^2 - 2x + \text{random noise}$', fontsize=16)
# Note: \text{} command is used to render "random noise" as regular text inside math mode.

plt.grid(True)
plt.legend()
plt.show()
```
